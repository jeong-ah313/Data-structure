## 문제1. 셀 정렬, 힙 정렬, 병합 정렬, 퀵 정렬
난수 발생기를 사용하여 배열에 정수 15개를 저장한다. 배열의 정수를 셸 정렬, 힙 정렬, 병합 정렬, 퀵 정렬하여 출력한다.

```python
import random
import copy

#셸정렬 알고리즘
def shellSort(A):
    n = len(A)
    gap = n // 2 #최초의 gap: 리스트의 크기의 절반
    while gap > 0:
        if (gap % 2) == 0:
            gap += 1 #gap이 짝수이면 1을 더함
        for i in range(gap):
            sortGapInsertion(A, i, n-1, gap)
        gap = gap//2
def sortGapInsertion(A, first, last, gap):
    for i in range(first+gap, last+1, gap):
        key = A[i]
        j = i - gap
        while j >= first and key<A[j]: #삽입위치를 찾음
            A[j+gap] = A[j]
            j = j - gap
        A[j + gap] = key #최종 위치에 삽입   
        
#힙정렬 알고리즘
#제자리 정렬로 구현한 힙 정렬
def heapify(arr, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    
    if l < n and arr[i] < arr[l]:
        largest = l
    if r < n and arr[largest] < arr[r]:
        largest = r
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest) #순환적으로 자식노드로 내려감
def heapSort(arr): #정렬되지 않은 배열-> 최대힙
    n = len(arr)
    for i in range(n//2, -1, -1): #최대힙 만들기
        heapify(arr, n, i)
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)
                
#병합정렬알고리즘
def mergeSort1(A):
    if len(A) > 1:
        mid = len(A) // 2
        left = A[:mid]
        right = A[mid:]
        mergeSort1(left)
        mergeSort1(right)
        merge1(A, left, right)
        
def merge1(A, left, right):
    i = 0
    j = 0
    k = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            A[k] = left[i]
            k, i = k + 1, i + 1
        else:
            A[k] = right[j]
            k, j = k + 1, j + 1
    while i < len(left):
        A[k] = left[i]
        k, i = k + 1, i + 1
    while j < len(right):
        A[k] = right[j]
        k, j = k + 1, j + 1

#퀵정렬 알고리즘
def quick_sort(arr):
    def sort(low, high):
        if high <= low:
            return

        mid = partition(low, high)
        sort(low, mid - 1)
        sort(mid, high)

    def partition(low, high):
        pivot = arr[(low + high) // 2]
        while low <= high:
            while arr[low] < pivot:
                low += 1
            while arr[high] > pivot:
                high -= 1
            if low <= high:
                arr[low], arr[high] = arr[high], arr[low]
                low, high = low + 1, high - 1
        return low

    return sort(0, len(arr) - 1)
        


def main():
    n = int(input("Enter a number of data:  "))
    data = []
    for i in range(n):
        numbers = random.randint(0, n)
        data.append(numbers)
        
    d = copy.deepcopy(data)
    print('Before shell sort:\t', d)
    shellSort(d)
    print('After shell sort:\t', d, '\n')
    
    d = copy.deepcopy(data)
    print('Before heap sort:\t', d)
    heapSort(d)
    print('After heap sort:\t', d, '\n')
    
    d = copy.deepcopy(data)
    print('Before merge sort:\t', d)
    mergeSort1(d)
    print('After merge sort:\t', d, '\n')
    
    d = copy.deepcopy(data)
    print('Before quick sort:\t', d)
    quick_sort(d)
    print('After quick sort:\t', d, '\n')
```

```python
main()
```

## 문제2
문제1의 각 정렬 방법마다 다음의 데이터 개수에 대하여 시간 측정을 하여 비교한다. python에서 제공하는 sort()함수를 사용하여 비교한다. 시간 측정은 실습8과 같은 방법으로 한다. 
데이터 개수: 100000, 500000, 1000000

```python
import time

##각 패스마다 배열에 저장된 데이터를 출력하지 않는 sort들로 재구성

while True:
    n = int(input("Enter a number of data: "))
    print()
    data = []
    for i in range(n):
        numbers = random.randint(0, n)
        data.append(numbers)
        
    if n == -1: break
        
    #shell Sort    
    d = copy.deepcopy(data) #정렬전 복사
    startTime = time.time()
    shellSort(d)
    endTime = time.time()
    print("Shell sort elapsed time      : %.3f seconds" %(endTime - startTime))
    
    #heap Sort
    d = copy.deepcopy(data) #정렬전 복사
    startTime = time.time()
    heapSort(d)
    endTime = time.time()
    print("Heap sort elapsed time       : %.3f seconds" %(endTime - startTime))
    
    #merge Sort
    d = copy.deepcopy(data) #정렬전 복사
    startTime = time.time()
    mergeSort1(d)
    endTime = time.time()
    print("Merge sort sort elapsed time : %.3f seconds" %(endTime - startTime))
    
    #quick Sort
    d = copy.deepcopy(data) #정렬전 복사
    startTime = time.time()
    quick_sort(d)
    endTime = time.time()
    print("Quick sort elapsed time      : %.3f seconds" %(endTime - startTime))
        
    #python의 sort()
    startTime = time.time()
    d.sort()
    endTime = time.time()
    print("Python sort elapsed time     : %.3f seconds" %(endTime - startTime), "\n")
```

```python
#+)prac
#퀵정렬 알고리즘
def quickSort(A, left, right):
    if left<right:
        q = partition(A, left, right)
        quickSort(A, left, q -1)
        quickSort(A, q + 1, right)
#분할함수(partition)
def partition(A, left, right):
    low = left + 1
    high = right
    pivot =  A[left]
    while (low <= high):
        while low <= right and A[low] < pivot:
            low += 1
        while high >= left and A[high] > pivot:
            high -= 1
        if low < high:
            A[low], A[high] = A[high], A[low]
            low, high = low + 1, high -1
        else:
            break
        A[left], A[high] = A[high], A[left]
        return high
```
