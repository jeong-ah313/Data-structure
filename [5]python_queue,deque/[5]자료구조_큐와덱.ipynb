## 1. 문제 1: 큐 구현 

큐를 클래스 사용하여 원형 큐로 구현한다. 클래스 큐를 다음과 같이 정의할 때 구현되지 않은 멤버 함수를 구현하고 실행한 결과와 같이 되도록 코딩한다. 큐의 초기 크기는 3으로 하고 (MAX_SIZE = 3) 큐가 가득 차면 큐의 크기를 2배로 늘려 데이터를 수용하도록 한다. 

```python
MAX_QSIZE = 3
class CircularQueue:
    def __init__(self): #생성자 함수
        self.items = [None] * MAX_QSIZE #멕스사이즈만큼 item를 생성
        self.front = 0
        self.rear = 0
        self.count = 0 #데이터 갯수
        self.maxSize = MAX_QSIZE #초기화
        
    #구현해야 할 멤버 함수
    def isEmpty(self):
        return self.front == self.rear
        
    def isFull(self):
        return self.front == (self.rear+1) % self.maxSize #안된다면 self빼기
    
    def clear(self): #현재데이터가 들어가 있음 모두 0으로
        self.front = 0
        self.rear = 0
        self.count = 0
        self.maxSize = MAX_QSIZE
        #self.front = self.rear
        
    def enqueue(self, item): #삽입연산
        if self.isFull(): #포화상태이면
            self.resize() #용량 늘리기
        
        if not self.isFull(): #포화상태가 아니면
            self.rear = (self.rear+1) % self.maxSize #rear 회전
            self.items[self.rear] = item #rear위치에 삽입
            self.count += 1
            
    def dequeue(self): #삭제연산
        if not self.isEmpty(): #공백상태가 아니면
            self.front = (self.front+1) % self.maxSize
            self.count -= 1
            return self.items[self.front]
            
    def resize(self): #크기 2배로(용량늘림)
        newitems = [None]*2*self.maxSize # 용량 x2
        scan = (self.front + 1) % self.maxSize
        for i in range(self.count): # 지금 현재 데이터만큼
            newitems[i+1] = self.items[scan] #newitems에 옮겨담기
            scan = (scan + 1) % self.maxSize #scan 숫자도 하나씩 증가
        self.items = newitems
        self.front = 0
        self.rear = self.count
        self.maxSize = 2*self.maxSize
        
    def peek(self): #맨앞에 있는 데이터 반환
        if not self.isEmpty():
            return self.items[(self.front + 1) % self.maxSize]
        
    def size(self):
        return self.count
    
    def print(self):
        out = []
        if self.front < self.rear:
            out = self.items[self.front + 1 : self.rear + 1]
        elif self.front > self.rear:
            out = self.items[self.front + 1 : self.maxSize] + self.items[0: self.rear + 1]
        print(out)
```

```python
def mylineEditor():
    queue = CircularQueue()
    print("Enter a command: e(nqueue), d(equeue), peek, s(ize), p(rint), or q(uit)")
    
    while True:
        command = input() 
    
        if command == 'e': #e(nqueue)
            item = input("추가할 data 입력")
            queue.enqueue(item)
            
        elif command == 'd': #dequeue
            print(queue.dequeue())
            
        elif command == 'peek':
            print(queue.peek())

        elif command == 's': #size
            print('size:', queue.size())

        elif command == 'p': #print
            queue.print()

        elif command == 'q':
            break
            
mylineEditor()
```

## 3. 문제 3: 덱 구현

문제 1에서 만든 원형 큐를 상속하여 덱을 구현한다. 클래스 덱을 다음과 같이 정의할 때 구현되지 않은 멤버 함수를 구현하고 실행한 결과와 같이 되도록 코딩 한다. 덱의 초기 크기는 3으로 하고 (MAX_SIZE = 3) 덱이 가득 차면 덱의 크기를 2배로 늘려 데이터를 수용하도록 한다. 

```python
MAX_QSIZE = 3
class CircularDeque(CircularQueue):
    def __init__(self):
        super().__init__() #생성자 함수는 super을 사용해서 호출해야한다.
        
    def addRear(self, item): #뒤에서 삽입
        self.enqueue(item) #삽입연산
    def deleteFront(self): #앞에서 삭제
        return self.dequeue()
    def getFront(self): #앞에 있는 데이터 반환
        return self.peek() #맨앞에서 반환
    
    def addFront(self, item): #앞에서 삽입
        if self.isFull(): #포화상태이면
            self.resize() #용량 늘리기
            
        if not self.isFull(): #포화상태가 아니면
            #self.front = (self.front+1) % self.maxSize #front 회전
            #self.enqueue(item) #뒤에서 삽입연산
            self.items[self.front] = item #항목저장 #front위치에 item이 삽입됨
            self.count += 1
            if self.isFull(): #front[0]위치에 삽입했으므로 포화상태일 것임
                self.resize() #따라서 용량늘려주기
            self.front = self.front -1 # 반시계방향 회전
            if self.front < 0: #front의 위치 바꿔주기
                self.front = self.maxSize - 1
                
    def deleteRear(self): #뒤에서 삭제
        if not self.isEmpty():
            #self.rear = (self.rear+1)%self.maxSize
            #self.count=-1
            #return self.items[self.rear]
            
            
            item = self.items[self.rear]
            self.rear = self.rear-1
            self.count -= 1
            if self.rear < 0:
                self.rear = self.maxSize -1
            return item
            
    def getRear(self): #뒤에꺼보기
        return self.items[(self.rear)]
```

```python
def mylineEditor2():
    Deque = CircularDeque()
    print("Enter a command: af(addFront), df(deleteFront), gf(getFront), s(ize) /\
            ar(addRear), dr(deleteRear), gr(getRear), p(rint), or q(uit)")
    
    while True:
        command = input() 
    
        if command == 'af': #앞에서 추가하기
            item = input("추가할 data 입력")
            Deque.addFront(item)
            
        elif command == 'df': #앞에서 삭제하기
            print(Deque.deleteFront())
            
        elif command == 'gf': #앞에서 보기
            print(Deque.getFront())

        elif command == 's': #size
            print('size:', Deque.size())
            
        elif command == 'ar': #뒤에서 추가하기
            item = input("추가할 data 입력")
            Deque.addRear(item)
            
        elif command == 'dr': #뒤에서 삭제하기
            print(Deque.deleteRear())
            
        elif command == 'gr': #뒤에꺼 보기
            print(Deque.getRear())

        elif command == 'p': #print
            Deque.print()

        elif command == 'q':
            break
            
mylineEditor2()
```

## 4. 문제 4: palindrome(회문) 테스트

Palindrome(회문)은 앞뒤 어느 쪽을 읽어도 같은 단어, 문장을 의미한다. 예를 들어 “level”, “2002”는 palindrome이다. 파일 “input1.txt”에 있는 문장을 읽어 palindrome인지 아닌지 판단하는 프로그램을 작성한다. Palindrome을 판단하는 방법은 여러가지 있으나 여기서는 한 줄의 각 문자를 덱에 저장한 후, 덱의 양쪽 끝에서 데이터의 중간까지 삭제하여 비교하는 방법을 사용한다. 덱은 문제 3에서 만든 것을 사용한다. 문장 중에 알파벳 문자는 대/소문자 구분하지 않고 검사한다

```python
Deque = CircularDeque()

one_line = open('input1.txt', 'r').read().splitlines() #한줄씩 문자열 읽기

for i in range(len(one_line)):
    print(one_line[i])
    one_line[i] = one_line[i].lower() #소문자로 변환
    word = list(one_line[i]) 
    
    for i in word:
        Deque.addRear(i)
        
    #하나의 단어에 대해서
    while Deque.count != 0: #덱의 사이즈가 비어있지 않을때
        if Deque.getRear() == Deque.getFront():#덱의 앞과 뒤가 같다면
            Deque.deleteRear() #뒤삭제
            Deque.deleteFront() #앞삭제
            if Deque.count == 0:
                print('  ==> palindrome')
            #else: continue
        else: 
            print('  ==> not palindrome')
            Deque.clear()
```
