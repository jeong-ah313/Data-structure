여기에 재귀적으로 호출할 때 공백을 주어 출력하면 위와 같은 모양이 되어 트리의 구조를 알 수 있게 된다.

```python
class BTNode:
    def __init__(self, item, left=None, right=None):
        self.item = item
        self.left = left
        self.right = right
        
class BTree:
    def __init__(self):
        self.root = None
    def buildTree(self):
        g = BTNode(12)
        d = BTNode(15, g)
        b = BTNode(10, None, d)
        e = BTNode(25)
        f = BTNode(55)
        c = BTNode(30, e, f)
        self.root = BTNode(20, b, c)
        
        
    ##구현해야 할 멤버 함수
    
    #중위순회
    def inorder(self): #맨 위 루트, 지금 self로 하나만 받음-> 이 self는 자기 자신을 순회하기 위함
        self.inorder1(self.root)
        print()
    def inorder1(self, node): #여기에 self.root가 들어옴
        if node is not None: #self.root는 None이 아니므로 돌아갈 것임
            self.inorder1(node.left) #왼쪽 서브트리 처리
            print(node.item, end=' ') #루트노드 처리(화면출력)
            self.inorder1(node.right) #오른쪽 서브트리 처리
    
    #전위순회
    def preorder(self):
        self.preorder1(self.root)
        print()
    def preorder1(self, node):
        if node is not None:
            print(node.item, end=' ') #루트노드 먼저 처리 및 출력
            self.preorder1(node.left) #왼쪽서브트리 처리
            self.preorder1(node.right) #오른쪽 서브트리 처리
    
    #후위순회
    def postorder(self): 
        self.postorder1(self.root)
        print()
    def postorder1(self, node): 
        if node is not None:
            self.postorder1(node.left)
            self.postorder1(node.right)
            print(node.item, end=' ')                      

    #노드 개수
    def nodeCount(self):
        return self.nodeCount1(self.root)
    def nodeCount1(self, node): #노드개수: 순환을 이용해서 트리의 노드 수를 계산하는 함수
        if node is None: #n이 None이면 공백 트리 --> 0반환
            return 0
        else: #좌우 서브트리의 노드 수의 합 +1을 반환(순환사용)
            return 1+self.nodeCount1(node.left)+self.nodeCount1(node.right) #왼쪽 서브트리 count하고 오른쪽 서브트리 count하고, 순환 이용=>모든 노드 개수 반환
    
    #단말노드 개수- 맨끝에 달린 노드개수
    def leafCount(self):
        return self.leafCount1(self.root)
    def leafCount1(self, node): #단말노드 개수
        if node is None: # 공백 트리 --> 0을 반환
            return 0
        elif node.left is None and node.right is None: #단말노드 -> 1을 반환
            return 1 #맨 위에 상위 루트 하나만 있는 경우
        else: #비단말 노드: 좌우 서브트리의 결과 합을 반환
            return self.leafCount1(node.left) + self.leafCount1(node.right)
        
    #트리 높이
    def height(self):
        return self.height1(self.root)
    def height1(self, node): #트리 높이
        if node is None: #공백 트리-> 0을 반환
            return 0
        hLeft = self.height1(node.left) #왼쪽 트리 높이
        hRight = self.height1(node.right) #오른쪽 트리 높이
        return max(hLeft,hRight) + 1
#         if (hLeft > hRight):
#             return hLeft + 1
#         else:
#             return hRight + 1
        
    def print(self):
        global node_h
        node_h = self.height1(self.root) #노드의 h값
        return self.RVL(self.root,0)
    
    #print출력을 위한 순회
    def RVL(self, node, depth):
        if node is not None:
            # 출력 문자열
            line = str(node.item)
            self.RVL(node.right, depth + 1)
            #왼쪽 자식이있으면 L 추가
            if(node.left is not None):
                line += ',L'
            #오른쪽 자식이있으면 R추가
            if(node.right is not None):
                line += ',R'
            # 공백을 10칸 주었으므로 문자열 길이 만큼 제외된 공백 뒤에 추가
            for i in range(10-len(line)):
                line += ' '    
            
            #깊이만큼 공백 추가 후 line 출력
            print(abs(depth)*'          ',line)
            
            self.RVL(node.left,depth + 1)

        
def main():
    bt = BTree()
    print("Enter a command: bt(build tree), inorder (traversal), pre(order traversal),")
    print("Enter(order traversal), nc(node count), lc(leaf count), h(eight),")
    print("pp(pretty print), or q(uit)")
    
    while True:
        print("> ", end = "")
        line = input().split()
        command = line[0]
        if command == 'bt':
            bt.buildTree()
        elif command == 'inorder':
            bt.inorder()
        elif command == 'pre':
            bt.preorder()
        elif command == 'post':
            bt.postorder()
        elif command == 'nc':
            print('node count = ', bt.nodeCount())
        elif command == 'lc':
            print("leaf count = ", bt.leafCount())
        elif command == 'h':
            print("height = ", bt.height())
        elif command == 'pp':
            bt.print()
        elif command == 'q':
            break
main()
```

#### 2. 문제 2: 우선 순위 큐

힙을 사용하여 우선 순위 큐를 구현한다. enqueue(), dequeue()를 실행하고 출력은 문제 1과 같이 시각적으로 힙의 모양을 알 수 있게 출력한다. 

우선 순위 큐를 사용하여 데이터를 정렬한다. 우선 순위 큐에서 dequeue()를 하면 제일 큰 값을 삭제하게 되며 이 데이터를 임시 리스트의 맨 끝에 저장하고, 다음에 dequeue()를 하여 리스트의 맨 끝 앞에 저장하며, 이런 식으로 데이터 개수만큼 dequeue()를 하여 리스트의 끝에서 앞으로 저장하면 정렬된 리스트를 얻게 된다. [주의: 파이썬 리스트의 sort()를 사용하지 말고 반드시 우선 순위 큐의 dequeue()를 사용하여 정렬한다.]

```python
class PriorityQueue:
    def __init__(self):
        self.items = []
        self.count = 0
    def size(self): return self.count
    def isEmpty(self): return self.count == 0
    #구현해야 할 멤버 함수
    def enqueue(self, item): #이진트리 성질 만족하기 위해, 맨끝에 대이터 삽입
        self.items.append(item)
        self.count += 1
        self.moveUp(0, self.count-1) #부모노드 비교-부모vs단말노드 비교해가면서 더 큰게 위로
    def moveUp(self, first, last):
        while last > first:
            parent = (last-1) // 2
            if self.items[parent] < self.items[last]:
                self.swap(parent, last) #swap
                last = parent
            else: break
    def swap(self, x, y):
        self.items[x], self.items[y] = self.items[y], self.items[x]
        
    def dequeue(self):
        if not self.isEmpty():
            item = self.items[0] #루트노드 임시 저장
            self.count -= 1
            self.items[0] = self.items[self.count] #맨 마지막 노드를 루트로 보냄
            self.moveDown(0, self.count-1)
            self.items.pop(-1) #맨 마지막 노드 삭제
            return item #items??????
    def moveDown(self, first, last):
        leftChild = 2*first + 1
        while leftChild <= last:
            if leftChild == last:
                maxChild =  leftChild
            else:
                rightChild = 2*first + 2
                if self.items[leftChild] <= self.items[rightChild]:
                    maxChild = rightChild
                else:
                    maxChild = leftChild
            if self.items[first] < self.items[maxChild]:
                self.swap(first, maxChild)
                first = maxChild
                leftChild = 2*first + 1
            else: break
        
    def print(self):
        print(self.items)
        
    def display(self):
        self.RVL(0, 0)
        
    def RVL(self, index, depth):
        if index < self.count:
            self.RVL(index *2 + 2, depth + 1)
            line = str(self.items[index])
            for i in range(10 - len(line)):
                line += ' '
            print(abs(depth)*'          ',line)
            self.RVL(index *2 + 1, depth + 1)
        
    def sort(self):
        result = []
        while self.isEmpty() == False:
            result.append(self.dequeue())
        for i in range(len(result)):
            self.items.append(result[len(result) - i - 1])
        self.count = len(self.items)
        return result
        
def main():
    pq = PriorityQueue()
    print("Enter a command = e(nqueue), d(equeue), empty, s(size)")
    print("p(rint), pp(pretty print), sort, or q(uit)")
    while True:
        print("> ", end = "")
        line = input().split()
        command = line[0]
        if command == 'e':
            item = int(line[1])
            pq.enqueue(item)
        elif command == 'd':
            print(pq.dequeue())
        elif command == 'empty':
            print(pq.isEmpty())
        elif command == 'p':
            pq.print()
        elif command == 'pp':
            pq.display()
        elif command == 's':
            print("size: ", pq.size())
        elif command == 'sort':
            pq.sort()
            pq.print()
        elif command == 'q':
            break
            
main()
```
