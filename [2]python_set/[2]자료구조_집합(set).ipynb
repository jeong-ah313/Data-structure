2. 리스트에 있는 데이터의 최솟값, 최댓값을 찾는 findMinMax() 함수를 공부하였다. 이 함수를 이용하여 파일 “input.txt”에 있는 연, 월, 일로 주어진 날짜를 읽어 출력하고 가장 오래된 날짜와 가장 늦은 날짜를 출력하도록 한다.

문제를 해결하기 위해 Date 클래스를 다음과 같이 정의하고 멤버 함수를 구현한다. 실행하기 위함 함수 main()은 주어진 파일을 읽어 날짜를 리스트에 저장한다. 날짜를 모두 읽은 후 리스트에 저장된 날짜를 출력하고 findMinMax() 함수 사용하여 리스트에 있는 날짜에서 가장 오래된 날짜와 가장 늦은 날짜를 출력한다. 

```python
class Date:
    def __init__(self,year,month,day):
        self.year = year
        self.month = month
        self.day = day
    def __gt__(self,rhs):
        value = Date(self.year, self.month, self.day)
        if self.year > rhs.year:
            return value
        elif self.year == rhs.year:
            if self.month > rhs.month:
                return value
            elif self.month == rhs.month:
                if self.day > rhs.day :
                    return value
                else:
                    return Date(rhs.year,rhs.month,rhs.day)
            else:
                return Date(rhs.year,rhs.month,rhs.day)
        else:
            return Date(rhs.year,rhs.month,rhs.day)
    def __lt__(self,rhs):
        value = Date(rhs.year,rhs.month,rhs.day)
        if self.year > rhs.year:
            return value
        elif self.year == rhs.year:
            if self.month > rhs.month:
                return value
            elif self.month == rhs.month:
                if self.day > rhs.day :
                    return value
                else:
                    return Date(self.year,self.month,self.day)
            else:
                return Date(self.year,self.month,self.day)
        else:
            return Date(self.year,self.month,self.day)
    def __str__(self):
        return str(self.year) + '/'+str(self.month) +'/' +str(self.day)

def findMinMax(lst):
    min_ = Date(9999,9999,9999)
    max_ = Date(-9999,-9999,-9999)
    for i in lst :
        max_ = max_.__gt__(i)
        min_ = min_.__lt__(i)
    return max_,min_
        
def main():
    inFile = open('input.txt','r')
    lst = []
    while True :
        line = inFile.readline()
        if line =="":
            break
        date = [int(i) for i in line.split()]
        lst.append(Date(date[0],date[1],date[2]))
    for i in range(len(lst)):
        print(lst[i])
    min_,max_ = findMinMax(lst)
    print()
    print('earlist date:',max_)
    print('latest date:',min_)
    inFile.close()

if __name__ == "__main__":
    main()
```

```python
class Date:
    def __init__(self,y,m,d):
        self.year = y
        self.month = m
        self.day = d
    def check_y(self):
        if self.year % 4 != 0 : return False
        if self.year % 100 != 0 : return True
        if self.year % 400 != 0 : return False
        return True
    def check_y_y(self):
        if self.check_y():
            return 366
        else:
            return 365
    def checkinmonth(self,year,month):
        if self.check_y():
            d = 29
        else:
            d = 28
        dic = {
            1:31,2:d,3:31,
            4:30,5:31,6:30,
            7:31,8:31,9:30,
            10:31,11:30,12:31
        }
        return dic[month]

    def increment_(self,element):
        day = self.day + element
        if day >= 365:
            while day >= 365:
                y_day = self.check_y_y()
                day -= y_day
                self.year +=1

        if day >= 29:
            while (day >= 60) and (day > 0):
                day -= self.checkinmonth(self.year,self.month)
                self.month +=1
                if self.month > 12 :
                    self.year +=1
                    self.month -= 12
            if self.month == 2:
                if (self.check_y) and day ==29:
                    pass
                elif (self.check_y) and day==30:
                    day -= 29
                    self.month += 1
                else:
                    day -= 28
                    self.month += 1

                self.day = day
            elif self.month in (1,3,5,7,8,10,12):
                if day > 31:
                    self.month += 1
                    day -= 31
                    self.day =day
                else:
                    self.day = day
            else:
                if day > 30 :
                    self.month +=1
                    day -= 30
                    self.day = day
                else:
                    self.day = day
        else:
            self.day = day
        if self.month > 12:
            self.month -= 12
            self.year +=1 
    def __str__(self):
        return str(self.year) + '/'+str(self.month) +'/' +str(self.day)

def main():
    lst = []
    with open('input2.txt','r') as f:
        while True :
            line = f.readline()
            if line =="":
                break
            data = [int(i) for i in line.split()]
            lst.append((Date(data[0],data[1],data[2]),data[3]))
        for i in lst:
            c,e = i
            print(c,"           "+str(e) + "일 후", " ==>",end="")
            c.increment_(e)
            print(c)
            

if __name__ == "__main__":
    main()
```

```python
import sys; sys.setrecursionlimit(10000)

def hanoi(n, fr, to, tmp):
    if n==1:
        print("원판1: %s --> %s" %(fr, to))
        return
    hanoi(n-1, fr, tmp, to)
    print("원판%d: %s --> %s" %(n, fr, to))
    hanoi(n-1, tmp, to, fr)

while(1):
    move=0
    count=int(input("옮길 원판의 갯수를 입력하시오 :"))
    move=(2**int(count))-1
    hanoi(count, 'A','C','B')
    print("\n이동횟수:",move,"\n")
    if count ==0:
        break
```
