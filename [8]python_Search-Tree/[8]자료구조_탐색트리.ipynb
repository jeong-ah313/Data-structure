#### 1. 문제 1: 정렬
주어진 데이터에 대해서 버블 정렬, 선택 정렬, 삽입 정렬을 사용하여 출력한다. 정렬 방법을 이해하기 위해서 각 패스마다 배열에 저장된 데이터를 출력한다.

데이터는 다음을 사용한다.
data = [24, 15, 29, 11, 47, 12]
실습 노트에 각 정렬 방법에 대해서 각 패스마다 정렬되는 과정을 연습한다.
버블 정렬, 선택 정렬, 삽입 정렬의 3개 정렬 방법의 실행 시간은 O(n2)이 되며 2개의 for 루프 사용하여 구현한다.
각 정렬 방법에 대해 같은 배열을 사용하기 위해서 정렬하기 전에 복사해서 사용한다. 
 import copy
 d = copy.deepcopy(data)

```python
import copy

data = [24, 15, 29, 11, 47, 12]

#중간과정 출력용 함수
def printStep(arr, val):
    print(" Pass %2d = " % val, end='')
    print(arr)


##버블정렬: 가장 작은 데이터를 왼쪽 끝으로
d = copy.deepcopy(data) #data는 그대로 두고 필요 시에 deepcopy하여 d에 저장하여 사용

print('Before sorting \n', d, '\n')
def bubbleSort(A):    
    n = len(A)
    for i in range(n-1):
        for j in range(n-1, i, -1):
            if A[j-1]> A[j]:
                A[j-1], A[j] = A[j], A[j-1]
        printStep(A, i+1)
        
print('After bubble sorting')
bubbleSort(d)
print('\n')
        
    
##선택정렬 알고리즘
d = copy.deepcopy(data)
print('Before sorting \n', d, '\n')
def selection_sort(A):
    n = len(A)
    for i in range(n-1):
        least = i
        for j in range(i+1, n):
            if (A[j]<A[least]):
                least = j
        A[i], A[least] = A[least], A[i] #교환
        printStep(A, i+1)
        
print('After selection sorting')
selection_sort(d)
print('\n')
        
    
##삽입 정렬 알고리즘
d = copy.deepcopy(data)
print('Before sorting \n', d, '\n')
def insert_sort(A):
    n = len(A)
    for i in range(1, n):
        key = A[i]
        j = i-1
        while j>=0 and A[j]>key:
            A[j+1]=A[j]
            j -= 1
        A[j+1] = key
        printStep(A, i)
        
print('After insertion sorting')  
insert_sort(d)     
```

#### 문제 3: 탐색

난수 발생기를 사용하여 배열에 정수를 저장한다. 배열을 deepcopy 하여 파이썬 리스트에서 제공하는 sort()를 사용하여 정렬한 후 임의의 숫자에 대해서 정렬되지 않은 데이터에 대해서는 순차 탐색과 정렬된 데이터에 대해서는 이진 탐색하여 시간 측정을 비교한다. 데이터를 찾으면 해당 인덱스를 출력한다.

```python
#난수 발생
n = int(input("Enter a number of data: "))
data = []
for i in range(n):
    numbers = random.randint(0, n)
    data.append(numbers) #저장
    
d = copy.deepcopy(data) #정렬전 복사

startTime = time.time()
d.sort() #python의 sort를 이용하여 copy
endTime = time.time()
print("Python sort elapsed time: %.3f seconds" %(endTime - startTime), '\n')


while True:
    n = int(input("Enter a number of data: "))

    if n == -1: break
        
    def sequential_search(A, target): #순차탐색
        low = 0
        high = len(A) - 1 
        for i in range(low, high+1):
            if A[i] == target :
                #print(n, 'is in the list at index', i)
                msg = ('%d is in the list at index %d' %(n, i))
                return msg
            
        #print(n, 'is not in the list') #탐색 실패시-> 얘도 무조건 출력 됨.
        msg = ('%d is not in the list' %n)
        return msg
    
    def binarySearch(A, target): #이진탐색
        low = 0 #맨 처음부터 시작해서
        high = len(A) - 1 #마지막값은  A의 길이에 -1을 하면 된다(마지막 값은 알아서 정렬 됨/앞의 값들이 다 정렬되어버려서)
        while low <= high:
            mid = (low + high) // 2
            if A[mid] == target: #중간값과 같으면, 그 값을 반환
                msg = ('%d is in the list at index %d' %(n, mid))
                #print(n, 'is in the list at index', mid)
                return msg
            elif target < A[mid]: #중간값보다 작으면, high값이 mid-1으로 설정되어 왼쪽에서 찾게 됨(low값은 그대로)
                high = mid - 1 #중앙값보다 크면, low값이 mid+1으로 설정되면서 오른쪽에서 찾게 됨(hight값은 그대로)
            else:
                low = mid + 1
        #print(n, 'is not in the list') #탐색 실패시
        msg = ('%d is not in the list' %n)
        return msg
    
    startTime = time.time()
    print(sequential_search(data, n)) #순차탐색 사용-정렬되지 않은 리스트
    endTime = time.time()
    print("Sequential search elapsed time: %.3f seconds" %(endTime - startTime))
    
    
    startTime = time.time()
    print(binarySearch(d, n)) # 이진탐색 사용-정렬된 리스트
    endTime = time.time()
    print("Binary search elapsed time: %.3f seconds" %(endTime - startTime), '\n')
    
```

#### 4. 문제 4: 해시 테이블을 체이닝 사용하여 구현

입력 파일 “student.txt”를 읽어 해시 테이블에 저장한다. “student.txt” 파일은 학생에 대한 정보가 저장되어 있다. 한 라인에 학번, 학생 이름, 성적으로 구성되어 있다. 해시 테이블에 저장할 때 엔트리는 학번을 키로, 학번, 학생 이름, 성적으로 구성된 레코드를 값으로 하여 저장한다. 충돌이 일어나면 체이닝 사용하여 연결한다. Student 클래스 만들어 사용한다. 삽입/탐색/삭제/출력에 대해서 구현한 해시 테이블을 테스트한다. 초기 테이블 크기는 5로 한다.

```python
class Student:
    def __init__(self, number, name, score):
        self.number = number
        self.name = name
        self.score = score
    def __str__(self):
        return '[' + str(self.number) + ', ' + self.name + ', ' + str(self.score) + ']'
    
class Node:
    def __init__(self, item, next = None):
        self.item = item
        self.next = next

class Entry:
    def __init__(self, key, value):
        self.key = key
        self.value = value
    def __str__(self):
        return str(self.key) + ":" + str(self.value)
    
class HashTable:
    def __init__(self, size):
        self.table = [None] * size
        self.tableSize = size
        
    def hash(self, key):
        sum = 0
        for ch in str(key):
            sum += ord(ch)
        return sum % self.tableSize
    
    #####구현해야 할 멤버 함수
    def insert(self, key, value): #삽입연산
        #체이닝을 이용한 해시 맵
        idx = self.hash(key) #해시 주소값 찾기(key에 의해)->교수님이 만든 함수이름은 hash, 교과서에는 hashFn
        self.table[idx] = Node(Entry(key, value), self.table[idx]) #전단삽입
        
        #entry = Entry(key, value) #1.엔트리 생성
        #node = Node(entry)        #2.엔트리로 노드 생성
        #node.link = self.table[idx] #3.노드의 링크필드 처리
        #self.table[idx] = node  #4.테이블의 idx 항목: node로 시작

    def search(self, key): #순차 탐색 연산
        idx = self.hash(key)
        node = self.table[idx]
        while node is not None:
            if node.item.key == key:
                return node.item
            node = node.next
        return None
        
    def delete(self, key): #삭제 연산: 항목 위치를 찾아 pop
        idx = self.hash(key)
        node = self.table[idx]
        before = None
        while node is not None:
            if node.item.key == key: 
                if before == None:
                    self.table[idx] = node.next
                else:
                    before.next = node.next
                return
            before = node #before갱신
            node = node.next #node갱신
            
    def print(self):
        for idx in range(len(self.table)):
            node = self.table[idx]
            #if node is not None:
            print("[%2d]"%idx, end='')
            while node is not None:
                print(node.item, end='\n    ') #연산자 중복함수(__str__)사용
                node = node.next
            print()
```

```python
def main():
    inFile = open("student.txt", "r")
    ht = HashTable(5) #초기 테이블크기 5
    while True:
        line = inFile.readline()
        if line == '':
            break
        lst = line.split()
        s = Student(int(lst[0]), lst[1], float(lst[2]))
        ht.insert(lst[0], s)
    ht.print()
    print("Enter a command: i(nsert), d(elete), s(earch), p(rint), or q(uit)")
    while True:
        command = input("> ")
        if command == 'i':
            print("Enter student number, name, score", end = '')
            line = input().split()
            s = Student(int(line[0]), line[1], float(line[2]))
            ht.insert(line[0], s)
        elif command == 'd':
            print("Enter student number: ", end = '')
            number = input()
            if ht.delete(number) is not None:
                print(str(number) + " not found") # not found
            else:
                print(str(number) + " deleted")
        elif command == 's':
            print("Enter student number: ", end = '')
            number = input()
            s = ht.search(number)
            if s is not None:
                print(s.value)
            else:
                print(str(number) + " not found")
        elif command == 'p':
            ht.print()
        elif command == 'q':
            break
main()
        
```
